```python
from tokenizers import Tokenizer
from tokenizers.models import Unigram
from tokenizers.trainers import UnigramTrainer
from tokenizers.pre_tokenizers import Whitespace


train_tokens = '/Users/benjaminbrooke/.cache/kagglehub/datasets/vadimkurochkin/wikitext-103/versions/1/wikitext-103/wiki.train.tokens'


tokenizer = Tokenizer(Unigram())
tokenizer.pre_tokenizer = Whitespace()

trainer = UnigramTrainer(
    vocab_size=16000,
    min_frequency=2,
    special_tokens=["[PAD]", "[CLS]", "[SEP]", "[MASK]"]
)

print("Training Unigram tokenizer...")
tokenizer.train([train_tokens], trainer)
print("Training complete.")


PAD_TOKEN = "[PAD]"
PAD_ID = tokenizer.token_to_id(PAD_TOKEN)

tokenizer.enable_padding(
    pad_id=PAD_ID,
    pad_token=PAD_TOKEN
)

tokenizer.enable_truncation(
    max_length=512
)
```

```python
def batch_iterator(path, batch_size=1024):
    batch = []
    with open(path, encoding="utf-8") as f:
        for line in f:
            batch.append(line.strip())
            if len(batch) == batch_size:
                yield batch
                batch = []
        if batch:
            yield batch
```

```python
from tqdm import tqdm
import torch

for batch in tqdm(batch_iterator(train_tokens, batch_size=1024)):

    encoded = tokenizer.encode_batch(batch)

    token_ids = [e.ids for e in encoded]
    attention_masks = [e.attention_mask for e in encoded]

    input_ids = torch.tensor(token_ids)
    attn_mask = torch.tensor(attention_masks)
```

```python
# tiny_transformer.py -- minimal Transformer encoder (1 layer)
import math
import torch
import torch.nn as nn
import torch.nn.functional as F

class Embedder(nn.Module):
    def __init__(self, vocab_size, d_model):
        super().__init__()
        self.tok_emb = nn.Embedding(vocab_size, d_model,padding_idx=0)
        self.pos_emb = nn.Embedding(5000, d_model)

    def forward(self, x):
        batch, seq_len = x.shape
        positions = torch.arange(seq_len, device=x.device)
        positions = positions.unsqueeze(0).expand(batch, seq_len)
        return self.tok_emb(x) + self.pos_emb(positions)
```

```python
class SelfAttention(nn.Module):
    def __init__(self, d_model, n_heads):
        super().__init__()
        self.n_heads = n_heads
        self.head_dim = d_model // n_heads

        self.q = nn.Linear(d_model, d_model)
        self.k = nn.Linear(d_model, d_model)
        self.v = nn.Linear(d_model, d_model)

        self.out = nn.Linear(d_model, d_model)

    def forward(self, x, attention_mask):
        B, T, C = x.shape

        q = self.q(x).view(B, T, self.n_heads, self.head_dim)
        k = self.k(x).view(B, T, self.n_heads, self.head_dim)
        v = self.v(x).view(B, T, self.n_heads, self.head_dim)

        q = q.transpose(1, 2)
        k = k.transpose(1, 2)
        v = v.transpose(1, 2)

        scores = q @ k.transpose(-2, -1)
        ####
        scores = scores / math.sqrt(self.head_dim)

        mask = attention_mask[:, None, None, :]
        scores = scores.masked_fill(mask == 0, float('-inf'))

        att = torch.softmax(scores, dim=-1)

        out = att @ v
        out = out.transpose(1, 2).contiguous()
        out = out.view(B, T, C)
        return self.out(out)

class FeedForward(nn.Module):
    def __init__(self, d_model):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(d_model, 4 * d_model),
            nn.ReLU(),
            nn.Linear(4 * d_model, d_model)
        )

    def forward(self, x):
        return self.net(x)


class TransformerBlock(nn.Module):
    def __init__(self, d_model, n_heads):
        super().__init__()
        self.attn = SelfAttention(d_model, n_heads)
        self.ff = FeedForward(d_model)
        self.ln1 = nn.LayerNorm(d_model)
        self.ln2 = nn.LayerNorm(d_model)

    def forward(self, x, attention_mask):
        # Self-attention + residual
        x = x + self.attn(self.ln1(x), attention_mask)

        # Feed-forward + residual
        x = x + self.ff(self.ln2(x))

        return x
```

```python
class TinyTransformer(nn.Module):
    def __init__(self, vocab_size, d_model=128, n_heads=4):
        super().__init__()
        self.embed = Embedder(vocab_size, d_model)
        self.block = TransformerBlock(d_model, n_heads)
        self.ln = nn.LayerNorm(d_model)
        self.out = nn.Linear(d_model, vocab_size)

    def forward(self, x, attention_mask=None):
        x = self.embed(x)
        x = self.block(x, attention_mask)
        x = self.ln(x)
        return self.out(x)
```

```python
Model_one = TinyTransformer(vocab_size=tokenizer.get_vocab_size(),d_model=64,n_heads=4)
```

```python
optimizer = torch.optim.Adam(Model_one.parameters(),lr=0.01)
```

```python
loss_fn = nn.CrossEntropyLoss(ignore_index=PAD_ID)
```

```python
from tqdm.notebook import tqdm

epochs = 1

batch_size = 256

for epoch in range(epochs):

    for batch in tqdm(batch_iterator(train_tokens, batch_size),desc="Training batches", unit="batch"):

            encoded = tokenizer.encode_batch(batch)

            input_ids = torch.tensor([e.ids for e in encoded])
            attention_mask = torch.tensor([e.attention_mask for e in encoded])

            inputs = input_ids[:, :-1]
            targets = input_ids[:, 1:]
            attn_mask = attention_mask[:, :-1]

            # Forward pass
            logits = Model_one(inputs, attn_mask)  # shape: (B, T, vocab_size)

            # Compute loss
            loss = loss_fn(
                logits.reshape(-1, logits.size(-1)),
                targets.reshape(-1)
            )

            # Backward pass
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
```
