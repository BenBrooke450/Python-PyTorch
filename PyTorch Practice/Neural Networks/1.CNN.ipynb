```python
import torch
from torch import nn

from torch.utils.data import DataLoader
from torchvision import transforms, datasets




Train_dataset = transforms.Compose([transforms.Resize((224,224)),transforms.ToTensor()])
Test_dataset = transforms.Compose([transforms.Resize((224,224)),transforms.ToTensor()])

train_dataset = datasets.ImageFolder("/Users/benjaminbrooke/.cache/kagglehub/datasets/aryashah2k/breast-ultrasound-images-dataset/versions/1/Dataset_BUSI_with_GT/Train_data",Train_dataset)
test_dataset = datasets.ImageFolder("/Users/benjaminbrooke/.cache/kagglehub/datasets/aryashah2k/breast-ultrasound-images-dataset/versions/1/Dataset_BUSI_with_GT/Test_data",Test_dataset)


train_loader = DataLoader(train_dataset,batch_size=1,shuffle=True)
test_loader = DataLoader(test_dataset,batch_size=1,shuffle=True)
```

```python
for X_val,y_val in train_loader:
    print(X_val,y_val)
```

```python
class MedicalImageModel(nn.Module):
    def __init__(self):
        super().__init__()

        self.convo_layer_one = nn.Conv2d(3,5,2)
        self.activation_one = nn.ReLU()
        self.pool_one = nn.MaxPool2d(kernel_size=4,stride=2)

        self.convo_layer_two = nn.Conv2d(5,10,2)
        self.activation_two = nn.ReLU()
        self.pool_two = nn.MaxPool2d(kernel_size=4,stride=2)

        self.convo_layer_three = nn.Conv2d(10,5,2)
        self.activation_three = nn.ReLU()
        self.pool_three = nn.MaxPool2d(kernel_size=4,stride=2)

        self.convo_layer_four = nn.Conv2d(5,3,2)
        self.activation_four = nn.ReLU()

        self.flatten =  nn.Flatten()

        self.linear_layer_one = nn.Linear(in_features=1728,out_features=32)
        self.activation_four = nn.ReLU()
        self.linear_layer_two = nn.Linear(in_features=32,out_features=3)

    def forward(self,x):
        x = self.pool_one(self.activation_one(self.convo_layer_one(x)))

        x = self.pool_two(self.activation_two(self.convo_layer_two(x)))

        x = self.pool_three(self.activation_three(self.convo_layer_three(x)))

        x = self.flatten(self.activation_four(self.convo_layer_four(x)))

        return self.linear_layer_two(self.activation_four(self.linear_layer_one(x)))
```

```python
model_one = MedicalImageModel()
```

```python
loss_fn = nn.CrossEntropyLoss()
```

```python
optimizer = torch.optim.Adam(model_one.parameters(),lr=0.01)
```

```python
y_loss = []
x_point = []
i = 0

epochs = 4
model_one.train()

for epoch in range(epochs):

    for x_batch, y_batch in train_loader:

        optimizer.zero_grad()

        y_pred = model_one(x_batch)

        loss = loss_fn(y_pred,y_batch)

        loss.backward()

        optimizer.step()



        x_point.append(i)
        y_loss.append(loss.item())

        i = i + 1

        if epoch % 1 == 0:
            print(f'Epoch {epoch}, Loss: {loss.item()}, y_pred_doftmax: {torch.softmax(y_pred, dim= 1).tolist()}')
```

```python
model_one.eval()

for x_eval, y_eval in test_loader:

    y_logits = model_one(x_eval)

    y_prob = torch.softmax(y_logits,dim=1)

    _, y_pred = torch.max(y_prob,dim = 1)

    one_hot = torch.nn.functional.one_hot(y_pred, num_classes=3)

    print(f"Probabilities: {y_prob.tolist()}, One hot:{one_hot},  Predictions {y_pred.item()}, Index of list True: {y_eval.item()}")
```
