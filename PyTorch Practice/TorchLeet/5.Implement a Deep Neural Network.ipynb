```python
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

print(iris)

# Features (sepal length, sepal width, petal length, petal width)
X = iris.data

# Target (species: setosa, versicolor, virginica)
y = iris.target
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np

pca = PCA(n_components=1)
X_reduced = pca.fit_transform(X)

print(X_reduced)

plt.scatter(X_reduced,y)
```

```python
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)
print(X_reduced,y)

# First principal component
X_reduced_one = X_reduced[:, 0]

# Second principal component
X_reduced_two = X_reduced[:, 1]


from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')  # create 3D axes

# scatter plot: x=PC1, y=PC2, z=PC3, color based on y (labels)
scatter = ax.scatter(X_reduced_one, X_reduced_two, y, cmap='viridis', s=50)
```

```python
import torch
from torch import nn


class DeepLearningNeuralNet(nn.Module):
    def __init__(self):
        super().__init__()

        self.linear_one = nn.Linear(in_features=4,out_features=30)
        self.linear_two = nn.Linear(in_features=30,out_features=10)
        self.linear_three = nn.Linear(in_features=10,out_features=3)

    def forward(self,x):
        return self.linear_three(self.linear_two(self.linear_one(x)))
```

```python
model_iris_one = DeepLearningNeuralNet()
```

```python
loss_fn = nn.CrossEntropyLoss()
```

```python
optimizer = torch.optim.SGD(model_iris_one.parameters(),lr=0.1)
```

```python
len(iris.data)
```

```python
from torch.utils.data import DataLoader, TensorDataset

X = iris.data
y = iris.target

indices = torch.randperm(len(X))

X = X[indices]
y = y[indices]

X_train = torch.tensor(X,dtype=torch.float32)[:100]
y_train = torch.tensor(y,dtype=torch.long)[:100]

X_test = torch.tensor(X,dtype=torch.float32)[100:]
y_test = torch.tensor(y,dtype=torch.long)[100:]

train_dataset = TensorDataset(X_train,y_train)
test_dataset = TensorDataset(X_test, y_test)

train_loader = DataLoader(train_dataset,batch_size=1)
test_loader = DataLoader(test_dataset,batch_size=1)
```

```python
print(X_train)
print(y_train)
```

```python
epochs = 3

losses = []
epochs_map = []
i = 0

for epoch in range(epochs):

    for epoch_X_train, epoch_y_train in train_loader:

        y_pred = model_iris_one(epoch_X_train)

        loss = loss_fn(y_pred,epoch_y_train)

        optimizer.zero_grad()

        loss.backward()

        optimizer.step()

        losses.append(loss.item())
        epochs_map.append(i)

        i = i + 1

        if epoch % 1 == 0:
            print(f'Epoch {epoch}, Loss: {loss.item()} batch {epoch_X_train}')
```

```python
plt.plot(epochs_map,losses)
```

```python
from torchmetrics import Accuracy

model_iris_one.eval()

epochs = 5

for epoch in range(epochs):

    for epoch_X_text, epoch_y_text in test_loader:

        y_pred_logits = model_iris_one(epoch_X_text)

        test_loss = loss_fn(y_pred_logits, epoch_y_text)

        y_pred_classes = torch.argmax(y_pred_logits, dim=1)

        acc = Accuracy(task="multiclass", num_classes=3)

        test_acc = acc(y_pred_classes, epoch_y_text)

        if epoch % 2 == 0:
            print(f"Epoch: {epoch}  Test Loss: {test_loss:.5f}, Test Accuracy: {test_acc:.2f}%, y_diff: {y_pred_classes.item(), epoch_y_text.item()}" )
```
