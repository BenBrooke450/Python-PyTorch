```python
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

print(iris)

# Features (sepal length, sepal width, petal length, petal width)
X = iris.data

# Target (species: setosa, versicolor, virginica)
y = iris.target
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np

pca = PCA(n_components=1)
X_reduced = pca.fit_transform(X)

print(X_reduced)

plt.scatter(X_reduced,y)
```

```python
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)
print(X_reduced,y)

# First principal component
X_reduced_one = X_reduced[:, 0]

# Second principal component
X_reduced_two = X_reduced[:, 1]


from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')  # create 3D axes

# scatter plot: x=PC1, y=PC2, z=PC3, color based on y (labels)
scatter = ax.scatter(X_reduced_one, X_reduced_two, y, cmap='viridis', s=50)
```

```python
import torch
from torch import nn


class DeepLearningNeuralNet(nn.Module):
    def __init__(self):
        super().__init__()

        self.linear_one = nn.Linear(in_features=4,out_features=30)
        self.linear_two = nn.Linear(in_features=30,out_features=10)
        self.linear_three = nn.Linear(in_features=10,out_features=3)

    def forward(self,x):
        return self.linear_three(self.linear_two(self.linear_one(x)))
```

```python
model_iris_one = DeepLearningNeuralNet()
```

```python
loss_fn = nn.CrossEntropyLoss()
```

```python
optimizer = torch.optim.SGD(model_iris_one.parameters(),lr=0.1)
```

```python

X = iris.data
y = iris.target

indices = torch.randperm(len(X))

X = X[indices]
y = y[indices]

X_train = torch.tensor(X,dtype=torch.float32)[:130]
y_train = torch.tensor(y,dtype=torch.long)[:130]

X_test = torch.tensor(X,dtype=torch.float32)[130:]
y_test = torch.tensor(y,dtype=torch.long)[130:]
```

```python
print(X_train)
print(y_train)
```

```python
epochs = 130

losses = []
epochs_map = []

for epoch in range(epochs):

    y_pred = model_iris_one(X_train)

    loss = loss_fn(y_pred,y_train)

    optimizer.zero_grad()

    loss.backward()

    optimizer.step()

    losses.append(loss.item())
    epochs_map.append(epoch)


    if epoch % 2 == 0:
        print(f'Epoch {epoch}, Loss: {loss.item()}')
```

```python
plt.plot(epochs_map,losses)
```
